# Capítulo 2

## Áreas entre curvas

Ahora vamos a usar la integral definida para calcular áreas limitadas por las gráficas de 2 funciones. Sean $f$ y $g$ 2 funciones continuas en el intervalo $[a, b]$ con la condición $f(x) \geq g(x)$ en $[a, b]$:

El área del rectangulo diferencial es:

**Altura**: $f(x_{i}^{*}) - g(x_{i}^{*})$, $x_{i}^{*} \in [x_{i-1}, x_{i}]$

**Grosor**: $\Delta x = \frac{b - a}{n}$

Por tanto,

$$
A \approx \sum_{i=1}^{n} [f(x_{i}^{*}) - g(x_{i}^{*})] \Delta x \Rightarrow A = \lim_{n \to \infty} \sum_{i = 1}^{n} [f(x_{i}^{*}) - g(x_{i}^{*})] \Delta x
$$

```python
# Define the interval [a, b]
a, b = 0, 5

# Define the functions f(x) and g(x)
def f(x):
    return np.sin(x) + 2

def g(x):
    return 0.5 * np.cos(2 * x) + 1

# Generate x values
x = np.linspace(a, b, 500)
y1 = f(x)
y2 = g(x)

# Choose a sample point x_i* and Δx
x_i_star = 2.5
delta_x = 0.3
height = f(x_i_star) - g(x_i_star)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the curves
ax.plot(x, y1, 'r', label='f(x) = sin(x) + 2')
ax.plot(x, y2, 'b', label='g(x) = 0.5*cos(2x) + 1')

# Fill the area between the curves
ax.fill_between(x, y1, y2, where=(y1 > y2), interpolate=True, color='gray', alpha=0.5, label='Área entre curvas')

# Draw the rectangle at x_i_star
rect_x = [x_i_star - delta_x/2, x_i_star + delta_x/2]
rect_y_bottom = g(x_i_star)
rect_y_top = f(x_i_star)
ax.bar(x_i_star, height, width=delta_x, bottom=rect_y_bottom, color='purple', alpha=0.7, edgecolor='black', label='Rectángulo en $x_i^*$')

# Mark a, x_i_star, and b
ax.axvline(a, color='black', linestyle='--')
ax.axvline(b, color='black', linestyle='--')
ax.axvline(x_i_star, color='green', linestyle='--', label='$x_i^*$')

# Add labels and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Área entre dos curvas y rectángulo en $x_i^*$')
ax.legend()

plt.show()
```

> **Definición**
>
> El **área A** de la región limitada por las curvas $y = f(x)$ y $y = g(x)$, y las rectas $x = a$, $x = b$, donde $f$, $g$ son funciones continuas y $f(x) \geq g(x)$, para todo $x \in [a, b]$, está dada por:
>
> $$A = \int_{a}^{b}{[f(x) - g(x)]}dx$$
